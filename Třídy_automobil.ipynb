```python
class Automobil:
  def __init__(self, typ, model, max_rychlost):
    self.typ = typ
    self.model = model
    self.max_rychlost = max_rychlost
    self.aktualni_rychlost = 0
    print(f"Auto {self.model} ({self.typ}) bylo vyrobeno.")

  def zrychli(self, o_kolik):
    nova_rychlost = self.aktualni_rychlost + o_kolik
    self.aktualni_rychlost = min(nova_rychlost, self.max_rychlost)
    print(f"{self.model}: Aktuální rychlost je {self.aktualni_rychlost} km/h")

  def brzdi(self, o_kolik):
    nova_rychlost = self.aktualni_rychlost - o_kolik
    self.aktualni_rychlost = max(nova_rychlost, 0)
    print(f"{self.model}: Auto zpomalilo na {self.aktualni_rychlost} km/h")

class NakladniAuto(Automobil):
  def __init__(self, model, max_rychlost, nosnost):
    super().__init__("Nákladní", model, max_rychlost)
    self.nosnost = nosnost
    print(f"Nosnost nákladního auta je {self.nosnost} tun.")

  def naloz_naklad(self, vaha):
    print(f"Nakládám {vaha} tun nákladu.")

class Elektromobil(Automobil):
  def __init__(self, model, max_rychlost, kapacita_baterie):
    super().__init__("Elektromobil", model, max_rychlost)
    self.kapacita_baterie = kapacita_baterie
    self.stav_baterie = 100
    print(f"Kapacita baterie je {self.kapacita_baterie} kWh.")

  def nabij(self):
    self.stav_baterie = 100
    print(f"Auto {self.model} je plně nabito.")

# podtřídy
tatra = NakladniAuto("Tatra Phoenix", 90, 20)
tesla = Elektromobil("Tesla Model S", 250, 100)

tesla.zrychli(100)
tatra.naloz_naklad(15)
```
